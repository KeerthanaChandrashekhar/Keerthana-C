```python
causes_for_diabetes = ['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age', 'diabetes']
causes_for_diabetes_counts = [
    len('pregnancies'),
    len('glucose'),
    len('diastolic'),
    len('triceps'),
    len('insulin'),
    len('bmi'),
    len('dpf'),
    len('age'),
    len('diabetes')
]

# Define colors for the pie chart # NEW
colors = ['blue', 'green', 'lightpink', 'yellow', 'coral','darkblue','brown','red','orange']

# Plot the pie chart
plt.figure(figsize=(10, 8))
plt.pie(causes_for_diabetes_counts, labels=causes_for_diabetes, autopct='%1.1f%%', startangle=140, colors=colors) # Pass the colors list to the colors parameter
plt.title('Distribution of Causes for Diabetes')
plt.axis('equal')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
